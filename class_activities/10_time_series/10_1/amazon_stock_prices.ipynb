# Time Series Basics

```python
import pandas as pd
from pathlib import Path
%matplotlib inline
```

```python
# Read the Amazon stock prices CSV file as a time series DataFrame 
# Be sure to set the `Date` column as the index
amz_df = pd.read_csv('./data/amazon.csv', parse_dates=True, index_col='Date')
amz_df.head()
```

### Slice Time Series Data

```python
# Select all rows from September 2018
sept_2018_df = amz_df.loc['2018-9']
sept_2018_df.head()
```

```python
# Select all rows from September through October for 2018
sept_to_oct_df = amz_df.loc['2018-9':'2018-10']
sept_to_oct_df
```

### Plot Time Series Data

```python
# Plot the closing prices using a line plot
amz_df['Close'].plot()
```

### Resample Time Series Data

```python
# Resample the closing prices to weekly and take the mean
weekly_mean_df = amz_df['Close'].resample('W').mean()
weekly_mean_df.head()
```

```python
# Plot the weekly average closing prices as a line chart
weekly_mean_df.plot()
```

# Optional Challenge

```python
# Use resample to get the highest closing price per month
monthly_max_df = amz_df['Close'].resample('M').max()

# Create a bar chart of the result
monthly_max_df.plot.bar()
```
